```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_theme()

%matplotlib inline
```

```python
reward_1 = np.load("separation_dfw_unequipped/reward.npy")
nmac_1 = np.load("separation_dfw_unequipped/nmacs.npy")
iterations_1 = np.load("separation_dfw_unequipped/iteration_record.npy")

reward_2 = np.load("separation_dfw_equipped2/reward.npy")
nmac_2 = np.load("separation_dfw_equipped2/nmacs.npy")
iterations_2 = np.load("separation_dfw_equipped2/iteration_record.npy")
```

```python
reward_df = pd.DataFrame(reward_1)
roll_mean = 10
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(reward_df.rolling(roll_mean).mean(),
        color="red")
# set x-axis label
ax.set_xlabel("Iteration", fontsize = 14)
# set y-axis label
ax.set_ylabel("Reward",
              color="red",
              fontsize=14)



ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.scatter(iterations_1, nmac_1,color="blue",marker="o")
ax2.set_ylabel("NMAC Count",color="blue",fontsize=14)

plt.show()

# save the plot as a file
# fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
#             format='jpeg',
#             dpi=100,
#             bbox_inches='tight')
```
